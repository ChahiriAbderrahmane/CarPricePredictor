Car_Production_Year

```python
df['Car Model'].astype(str)

for row in df['Car Model']:
    df['Year'] = df['Car Model'].str.extract(r"(\d{4})") 
df['Year'] = pd.to_numeric(df['Year'])
```

Car_Brand_name

```python
car_brands = ["Oldsmobile Cutlass","Saab","Lamborghini","Chrysler","Bentley","Buick","GMC","Acura","Cadillac","BMW","Mercedes-Benz","Audi","Lexus","Ford","Hyundai","Kia","Honda","Tesla","Ferrari","Porsche","CitroÃ«n","Peugeot","Renault","Mazda","Mitsubishi","Subaru","Alfa Romeo","Fiat","Maserati","Chevrolet","Dodge","Jeep","Volkswagen","Nissan","Toyota","Volvo","SsangYong","BYD","Great Wall","Aston Martin","Nissan","Oldsmo_bile","Packard","Pagani","Panoz","Plymouth","Polestar","Pontiac","Porsche","RAM","Ram_bler","Renault","Rivian","Rolls-Royce","Saa_b","Saturn","Scion","smart","Stude_baker","Su_baru","Sun_beam","Suzuki","Tesla","Toyota","Triumph","VinFast","Volkswagen","Volvo","Willys","Kia","Koenigsegg","Lam_borghini","Lancia","Land Rover","Lexus","Lincoln","Lotus","Lucid","Maserati","May_bach","Mazda","McLaren","Mercedes-_benz","Mercury","MG","MINI","Mitsu_bishi","Nash","Fisker","Ford","GAZ","Genesis","Geo","GMc","Honda","Hudson","Hummer","Hyundai","INEOS","INFINITI","International","Isuzu","Jaguar","Jeep","Jensen","Kaiser","Karma","_buick","cadillac","checker","chevrolet","chrysler","citroen","cord","Datsun","Delahaye","Delorean","Desoto","DeTomaso","Dodge","Eagle","Edsel","Excali_bur","Facel-Vega","Ferrari","FIAT"]
len(car_brands)

def brand(row):
    for brand in car_brands:
        if brand in row:
            return brand
df['brand'] = df['Car Model'].apply(brand)
```

Mileage_in_mile_unit

```python
df['Mileage'] = df['Mileage'].str.replace(' mi.','',regex=False).str.replace(',','',regex=False)
df['Mileage_mi'] = pd.to_numeric(df['Mileage'], errors='coerce')
```

Price_cleaning

```python
df['Price (USD)'].astype(str)
df['Price (USD)'] = df['Price (USD)'].str.replace('$','',regex=False).str.replace(',','',regex=False)
df['Price (USD)'] = pd.to_numeric(df['Price (USD)'], errors='coerce')
```

```python
df['Price (USD)'].isnull().sum()
```

```python
df = df[ df['Price (USD)'].notnull() ]
```

Handeling_Car_Condition

```python
df['Condition'].unique()
```

```python
df[df['Condition'] == 'Prequalify now']
```

```python
df[ df['Condition'] == 'New & Used' ]
```

```python
df['Condition_car'] = None

for index, row in df.iterrows():
    if row['Condition'] == 'Used' or row['Condition'] == 'Prequalify now':
        df.loc[index, 'Condition_car'] = 'Used'
    elif row['Condition'] == 'New' or row['Condition'] == 'New & Used':
        df.loc[index, 'Condition_car'] = 'New'
    elif pd.notna(row['Condition']) and 'Certified' in row['Condition']:
        df.loc[index, 'Condition_car'] = 'Certified'
```

```python
df['Condition_car'].unique()
```

```python
df['Condition_car'].isnull().sum()
```

```python
df
```

```python
for index, row in df.iterrows():
    # Check if 'Condition' is an empty string or NaN using pd.isnull() after converting it to a string
    if pd.isnull(row['Condition']) and not pd.isnull(row['Mileage_mi']):  
        df.loc[index, 'Condition_car'] = 'Used'
    # Check if 'Condition' is an empty string or NaN using pd.isnull() after converting it to a string and if 'Mileage' is NaN
    elif pd.isnull(row['Condition']) and pd.isnull(row['Mileage_mi']):  
        df.loc[index, 'Condition_car'] = 'New'
```

```python
df = df[df['Condition_car'].notnull()]
```

```python
df['Condition_car'].isnull().sum()
```

Monthly_Payment_Cleaning

```python
df['Monthly Payment'] = df['Monthly Payment'].astype(str).str.extract(r'(\d[\d,]*)').replace(',','',regex=True).astype(float)
```

```python
df['Monthly Payment'] = pd.to_numeric(df['Monthly Payment'], errors='coerce')
```

```python
df['Monthly Payment'].isnull().sum()
```

Final noticements
